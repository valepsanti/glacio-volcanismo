# Evolución de los discos de hielo
## Modelo de deglaciación cíclica (M2)

# 5 % y cada 1000 años

```python
import pandas as pd #visualizar y trabajar con datos
import os #acceder a funcionalidades dependientes del Sistema Operativo\
import numpy as np
import matplotlib.pyplot as plt #visualizar datos (graficos, mapas...)
%matplotlib inline
import math #utilizar fórmulas matemáticas (raiz, pi...)

def save_excel_file(df, filename):
    if os.path.isfile(filename):
        print(f"El archivo '{filename}' ya existe aquí")
    else:
        df.to_excel(filename)
        print(f"El archivo '{filename}' se ha guardado")



M1_Tabla_CVMCh = pd.read_excel("Input//M1_CVMCh.xlsx")
M1_Tabla_VH = pd.read_excel("Input//M1_VH.xlsx")

DAVIES_Tabla_CVMCh = pd.read_excel("Input//Tabla_radiosfinal_CVMCh.xlsx")
DAVIES_Tabla_VH = pd.read_excel("Input//Tabla_radiosfinal_VH.xlsx")

t_cvmch_D =DAVIES_Tabla_CVMCh['tiempo']
r_cvmch_D = DAVIES_Tabla_CVMCh['Radio (km)']

t_vh_D =DAVIES_Tabla_VH['tiempo']
r_vh_D = DAVIES_Tabla_VH['Radio (km)']

############################################
############################################
########  DEFINICIÓN DE CICLOS  ############
############################################
############################################
reduccion = 0.05
delta_t = 1000
porcen = round(reduccion * 100)

#####################################
###########  CVMCh ###############
################################

tiempo = M1_Tabla_CVMCh['tiempo']
radio = M1_Tabla_CVMCh['Rt']

modelo2_radio_mc = [(r * (1 - reduccion * abs(math.sin(t * math.pi / delta_t)))) for t, r in zip(tiempo, radio)]
superficies_mc = [(r ** 2) * math.pi for r in modelo2_radio_mc]

k = 0.033
y = 1.36

volumen_mc = [k * (A ** (y - 1)) * A for A in superficies_mc]
h_mc = [v / A for v, A in zip(volumen_mc, superficies_mc)]

tabla_m2_05_CVMCh = pd.DataFrame({
    "volcan": 'CVMCh',
    "tiempo": tiempo,
    "edad": M1_Tabla_CVMCh['edad'],
    "Rt": modelo2_radio_mc,
    "ht": h_mc,
    "At": superficies_mc,
    "Vt": volumen_mc
})

#####################################
###########  V H ###############
################################

tiempo = M1_Tabla_VH['tiempo']
radio = M1_Tabla_VH['Rt']

modelo2_radio_vh = [(r * (1 - reduccion * abs(math.sin(t * math.pi / delta_t)))) for t, r in zip(tiempo, radio)]
superficies_vh = [(r ** 2) * math.pi for r in modelo2_radio_vh]

k = 0.033
y = 1.36

volumen_vh = [k * (A ** (y - 1)) * A for A in superficies_vh]
h_vh = [v / A for v, A in zip(volumen_vh, superficies_vh)]

tabla_m2_05_VH = pd.DataFrame({
    "volcan": 'VH',
    "tiempo": tiempo,
    "edad": M1_Tabla_VH['edad'],
    "Rt": modelo2_radio_vh,
    "ht": h_vh,
    "At": superficies_vh,
    "Vt": volumen_vh
})
# Guardar el archivo para CVMCh
filename_cvmch = "Output/M2-5_1000_CVMCh.xlsx"
save_excel_file(tabla_m2_05_CVMCh, filename_cvmch)

# Guardar el archivo para VH
filename_vh = "Output/M2-5_1000_VH.xlsx"
save_excel_file(tabla_m2_05_VH, filename_vh)
####################################################################################################################

fig, ax1  = plt.subplots(figsize=(10, 5))

ax1.plot(tiempo,modelo2_radio_mc,label="Modelo 2 CVMCh")
ax1.plot(tiempo,modelo2_radio_vh,label="Modelo 2 VH")
ax1.plot(tiempo,M1_Tabla_CVMCh['Rt'],label="Modelo 1 CVMCh")
ax1.plot(tiempo,M1_Tabla_VH['Rt'],label="Modelo 1 VH")
ax1.scatter(t_cvmch_D,r_cvmch_D,label="Davies el al.,2020",color = 'purple')
ax1.scatter(t_vh_D,r_vh_D,color = 'purple')

ax1.set_ylim(0,180)
ax1.set_xlim(0,35300)
ax1.grid(b=True)

ax1.set_title(f"Modelo cíclico reducción del {porcen}% ciclo {delta_t} años ") 
ax1.legend(loc='upper right')

ax1.set_ylabel("Radio del disco de hielo (km)", fontsize=10)
ax1.set_xlabel("tiempo (años)", fontsize=10)

ax1.set_ylim(0,80)
ax1.set_xlim(0,35500)

#__

a = [35000,34800, 30000, 25000,22000,20000,15000,10000,5000,0]
ax2 = ax1.twiny()
#ax2.scatter(a,Tabla_CVMCh['Radio (km)'])
ax2.invert_xaxis()
ax2.set_xlim (35.0,-0.5)

ax1.axvline(x= 35000,color="grey", linestyle="--",linewidth=1)

ax2.set_xlabel("Edad (ka)", labelpad=5, fontsize=10)

ax1.grid(linewidth=0.75, color='grey', linestyle='dotted')

from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)

xmajor_tick = np.arange(0, 35500, 5000)
xminor_tick = np.arange(0, 35200, 500)
ax1.set_xticks(xmajor_tick)
ax1.set_xticks(xminor_tick, minor=True)

ymajor_tick = np.arange(0, 181, 20)
yminor_tick = np.arange(0, 180, 5)
ax1.set_yticks(ymajor_tick)
ax1.set_yticks(yminor_tick, minor=True)

xmajor_tick = np.arange(0, 35.500, 5.000)
xminor_tick = np.arange(0, 35.200, .500)
ax2.set_xticks(xmajor_tick)
ax2.set_xticks(xminor_tick, minor=True)
     
fig.tight_layout()
```
