# SELECCIÓN Y ANÁLISIS DE DATOS BIBLIOGRÁFICOS

## Paso 1: Visualizar Datos

```python
import pandas as pd #visualizar y trabajar con datos
import os #acceder a funcionalidades dependientes del Sistema Operativo\
import geopandas as gpd #trabajar con sistema de coordenadas
import numpy as np
import matplotlib.pyplot as plt #visualizar datos (graficos, mapas...)
%matplotlib inline
from shapely.geometry import Polygon #trabajar con objetos geométricos
import math #utiluzar fórmulas matemáticas (raiz, pi...)
import imageio ## mostrar gif


shp_volcanes = gpd.read_file("Input//Volcanes")
chile_clip = gpd.read_file("Input//Chile_clip")


ages = ["35","30","25","20","15","13","10","5","02","0"] 
icesheet_ages = [] 
for age in ages: 
    icesheet = gpd.read_file(f"Input//PATICE_icesheets//Icesheet{age}ka.shp")
    if icesheet.crs == 'epsg:32718':
        icesheet_ages.append(icesheet) 
    else:
        sheet = icesheet.to_crs('epsg:32718')
        icesheet_ages.append(sheet) 

# Crear una figura y una matriz de ejes de 2 filas y 5 columnas
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(10, 6))

# Iterar sobre la matriz de ejes y la lista de marcos de datos
for i, ax in enumerate(axes.flatten()):
    base = chile_clip.plot(color="white", edgecolor="black", ax=ax)
    for j, icesheet in enumerate(icesheet_ages):
        # Utilizar el marco de datos correspondiente de icesheet_ages
        if j == i:
            icesheet.plot(ax=base, color="blue", edgecolor="black")
            shp_volcanes.plot(ax=base, color="r")
            
    ax.set_ylim (4700000,5800000)
    ax.set_xlim (420000,980000)
            


    ax.set_title(f"IceSheet {ages[i]} ka")

plt.tight_layout()
plt.show()
```

## Paso 2a: Delimitación de áreas de estudio

SE UTILIZAN LOS RADIOS DE LA MÁXIMA EXTENSIÓN GLACIAR (MEG) EN CADA TIEMPO (definidos en ARCGIS)

```python
radio_meg = pd.read_excel("Input//superficie_MEG.xlsx")

volcanes = ['VH', 'CVMCh'] # Lista con los nombres de los volcanes

# Crear dos DataFrames para almacenar los resultados
radios_VH = pd.DataFrame(columns=radio_meg.columns)
radios_CVMCh = pd.DataFrame(columns=radio_meg.columns)

for volcan in volcanes:
    radio = radio_meg.loc[radio_meg['volcan'] == volcan]
    if volcan == 'VH':
        radios_VH = radios_VH.append(radio)
    else:
        radios_CVMCh = radios_CVMCh.append(radio)

# Imprimir los resultados
print("Radios del volcán VH:")
print(radios_VH)
print("\nRadios del volcán CVMCh:")
print(radios_CVMCh)

### Separar volcans  sin perder la caract. de geopandas
shp_cvmch = shp_volcanes.loc[[0],'geometry'] 
shp_vh = shp_volcanes.loc[[1],'geometry']

def crear_circulo(volcan_nombre, radio_MEG_volcan, volcan_shp):
    r_MEG = []
    for r in radio_MEG_volcan*1000:   ##se cambia la unidad de medida de km a m, porque así funciona "geometry"
        r_vn = volcan_shp.buffer(r)
        circunferencia = gpd.GeoDataFrame(geometry=r_vn)
        circunferencia["volcan"] = volcan_nombre
        r_MEG.append(circunferencia)
    return r_MEG

rMEG_CVMCh = crear_circulo("CVMCh", radios_CVMCh['RMEG'], shp_cvmch)
rMEG_VH = crear_circulo("VH", radios_VH['RMEG'], shp_vh)

############################################################################
###########################################################################
####VISUALIZAR DATOS#####

# Crear una figura y una matriz de ejes de 2 filas y 5 columnas
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(10, 6))

# Iterar sobre la matriz de ejes y las listas de radios correspondientes a cada volcán
for i, ax in enumerate(axes.flatten()):
    base = chile_clip.plot(color="white", edgecolor="black", ax=ax)
    r_vh = rMEG_VH[i]
    r_cvmch = rMEG_CVMCh[i]
    if i<4:  ### por la falta de radios en el CVMCh desde los 15ka
        r_cvmch.plot(ax=base, color="blue", edgecolor="black")
        r_vh.plot(ax=base, color="red", edgecolor="black")
    else:
        r_vh.plot(ax=base, color="red", edgecolor="black")

    ax.set_ylim (4700000,5800000)
    ax.set_xlim (420000,980000)
    ax.set_title(f"IceSheet {ages[i]} ka")

plt.tight_layout()
plt.show()

```

## Paso 2B: Delimitación de áreas de estudio

SE REALIZA UN INTERSECCIÓN ENTRE LOS MEG DE CADA VOLCAN Y LOS ICESHEET, PARA OBTENER LOS NUEVOS RADIOS HIPOTETICOS

```python
base_corte_cvmch = rMEG_CVMCh # input
base_corte_vh = rMEG_VH # input
shp_a_cortar = icesheet_ages # los IceSheets en una lista (35ka,30ka,25ka,20ka)

##existen algunas inconsistencia en las tablas (columna ages)

ages = [35000, 30000, 25000, 20000, 15000, 13000, 10000, 5000,200,0]
for i, age in enumerate(ages):
    icesheet_ages[i]["Age"] = age
    
##################################################################
# interceptar shpfiles, la intersection entraga el calculo de los pedazos recortados 
    
def calcular_area_clip(shp_a_cortar, base_corte):
    shp_radius = []
    for circulo, shpfile in zip(base_corte, shp_a_cortar):
        shp_cortado = gpd.overlay(circulo, shpfile, how="intersection", make_valid=True)
        shp_cortado["Area_Clip_km2"] = shp_cortado.area/1000000  ## volver a km2
        shp_radius.append(shp_cortado[["volcan", "Age", "Area_Clip_km2"]])
    return pd.concat(shp_radius)

#### se suma todos los pedazos

Tabla_rfinal_Area_total_cvmch = calcular_area_clip(shp_a_cortar, base_corte_cvmch).groupby(["volcan", "Age"], as_index=False)["Area_Clip_km2"].sum()

Tabla_rfinal_Area_total_vh = calcular_area_clip(shp_a_cortar, base_corte_vh).groupby(["volcan", "Age"], as_index=False)["Area_Clip_km2"].sum()

##################################3
# Calculo de los nuevos radios a partir de el área total

radio_hipo_cvmch = []
for area in Tabla_rfinal_Area_total_cvmch["Area_Clip_km2"]:
    radio = (math.sqrt(area/math.pi))
    radio_hipo_cvmch.append(radio)
    
Tabla_rfinal_Area_total_cvmch["Radio_Hipo_km"] = radio_hipo_cvmch

radio_hipo_vh = []
for area in Tabla_rfinal_Area_total_vh["Area_Clip_km2"]:
    radio = (math.sqrt(area/math.pi))
    radio_hipo_vh.append(radio)
    
### completar tabla de CVMCh
    
Tabla_rfinal_Area_total_vh["Radio_Hipo_km"] = radio_hipo_vh

df = pd.DataFrame({'volcan': ['CVMCh','CVMCh','CVMCh','CVMCh','CVMCh','CVMCh'],
           'Age': [0,200,5000,10000,13000,15000],
           'Area_Clip_km2':['0','0','0','0','0','0'],
           'Radio_Hipo_km':['0','0','0','0','0','0']})

Tabla_rfinal_Area_total_cvmch= pd.concat([df,Tabla_rfinal_Area_total_cvmch], ignore_index=True)

print('***********************************************')
print("Nuevos RADIOS del volcán CVMCh:")
print(Tabla_rfinal_Area_total_cvmch)
print('***********************************************')
print("Nuevos RADIOS del volcán VH:")
print(Tabla_rfinal_Area_total_vh)
print('***********************************************')
```

```python
### RELLENAR DATOS CON BIBLIOGRAFIA PARA EL PRESENTE

ACVMCh = 15.1 #bibliografía  Scheiter M. et al., 2021
rcvmch = math.sqrt(ACVMCh/math.pi)

Tabla_rfinal_Area_total_cvmch.loc[0]=["CVMCh",0,ACVMCh,rcvmch]
Tabla_rfinal_Area_total_cvmch

AVH = 69.84 #bibliografía Rivera A. et al., 2017
rvh = math.sqrt(AVH/math.pi)

Tabla_rfinal_Area_total_vh.loc[0]=["VH",0,AVH,rvh]
Tabla_rfinal_Area_total_vh
print("*********************************************")
print("nombre variante = 'Tabla_rfinal_Area_total_cvmch' ")
print("*********************************************")
print("RESULTADOS CVMCh")
print(Tabla_rfinal_Area_total_cvmch)
print("*********************************************")
print("nombre variante = 'Tabla_rfinal_Area_total_vh' ")
print("*********************************************")
print("RESULTADOS VH")
print(Tabla_rfinal_Area_total_vh)
print("*********************************************")
print("-----------------------------------------------------------------------------------------")

```

PARA COMPLETAR LOS DATOS QUE FALTAN PARA EL CVMCh, SE UTILIZA COMO MODELO EL VH
PRIMERO SE CALCULA EL % DE REDUCCIÓN DEL ICESHEET EN EL TIEMPO PARA CADA VOLCAN


```python

###DISMINUCION
R_inicial_VH = float(Tabla_rfinal_Area_total_vh['Radio_Hipo_km'].loc[9])  ##Seleccion del radio inicial (a los 35ka) para el VH
R_inicial_CVMCh = float(Tabla_rfinal_Area_total_cvmch['Radio_Hipo_km'].loc[9]) ##Seleccion del radio inicial (a los 35ka) para el  CVMCh

Reduccion_RVH = []
Reduccion_RCVMCh = []
for i in range(10):
    rc= 100 * float(Tabla_rfinal_Area_total_cvmch['Radio_Hipo_km'].loc[i])/R_inicial_CVMCh  #hay que modificar el valor en 0k
    rv= 100 * float(Tabla_rfinal_Area_total_vh['Radio_Hipo_km'].loc[i])/R_inicial_VH
    Reduccion_RVH.append(rv)
    Reduccion_RCVMCh.append(rc)



Reduc_Rcvmch = pd.DataFrame({'volcan': ['CVMCh','CVMCh','CVMCh','CVMCh','CVMCh','CVMCh','CVMCh','CVMCh','CVMCh','CVMCh'],
           'Age': [0,200,5000,10000,13000,15000,20000,25000,30000,35000],
           '% de reducción': Reduccion_RCVMCh })

Reduc_Rvh = pd.DataFrame({'volcan': ['VH','VH','VH','VH','VH','VH','VH','VH','VH','VH'],
           'Age': [0,200,5000,10000,13000,15000,20000,25000,30000,35000],
           '% de reducción': Reduccion_RVH })


###################################################################################################
#####################################################################
#### SE PROCEDE A COMPLETAR DATOS DEL CVMCh CON LOS DEL VH

Reduc_Rcvmch["% de reducción"].loc[1:5] = list(Reduc_Rvh["% de reducción"].loc[1:5])

#####################################################################################

print("*********************************************")
print("nombre variante = 'Reduc_Rcvmch' ")
print("***********************************")
print("RESULTADOS CVMCh")
print(Reduc_Rcvmch)
print("***********************************")
print("nombre variante = 'Reduc_Rvh' ")
print("***********************************")
print("RESULTADOS VH")
print(Reduc_Rvh)
print("***********************************")
```

SE CALCULAN LOS NUEVOS RADIOS PARA EL CVMCH QUE COMPLETAN EL REGISTRO HASTA EL PRESENTE

```python
#################################################################3
##NUEVOS RADIOS PARA CVMCH
    
Rt_CVMCh =[]
for i in range(10):
    R = R_inicial_CVMCh*Reduc_Rcvmch["% de reducción"].loc[i]/100
    Rt_CVMCh.append(R)
    
Tabla_Rt_CVMCh = pd.DataFrame({'volcan': ['CVMCh','CVMCh','CVMCh','CVMCh','CVMCh','CVMCh','CVMCh','CVMCh','CVMCh','CVMCh'],
           'edad': [0,200,5000,10000,13000,15000,20000,25000,30000,35000],
           'Superficie Glacial (km2)' : [(r**2) * math.pi for r in Rt_CVMCh] ,             
           'Radio (km)': Rt_CVMCh })

R_HIPOVH = Tabla_rfinal_Area_total_vh['Radio_Hipo_km']

Tabla_Rt_VH = pd.DataFrame({'volcan': ['VH','VH','VH','VH','VH','VH','VH','VH','VH','VH'],
           'edad': [0,200,5000,10000,13000,15000,20000,25000,30000,35000],
           'Superficie Glacial (km2)' : [(r**2) * math.pi for r in R_HIPOVH]  ,          
           'Radio (km)': R_HIPOVH })

#################################################################3
##AGREGAR ESPESOR Y VOLUMEN

k = 0.033 #constante usada en ecuación 
y = 1.36 #valor definido para ice field y glaciares (DAVIES ET AL., 2020)

Espesor_VH = []
for A in Tabla_Rt_VH['Superficie Glacial (km2)']:
    h = k * (A)**(y-1)
    Espesor_VH.append(h)
    
Espesor_CVMCh = []
for A in Tabla_Rt_CVMCh['Superficie Glacial (km2)']:
    h = k * (A)**(y-1)
    Espesor_CVMCh.append(h)
                            
Tabla_Rt_CVMCh ['Espesor (m)'] =[( h*1000) for h in Espesor_CVMCh]
Tabla_Rt_VH ['Espesor (m)'] =[( h*1000) for h in Espesor_VH]

tuplaCVMCh  = list(zip(Tabla_Rt_CVMCh ['Superficie Glacial (km2)'],Espesor_CVMCh))
tuplavh  = list(zip(Tabla_Rt_VH ['Superficie Glacial (km2)'],Espesor_VH ))
volumen_cvmch = []
volumen_vh = []

for i in range(10):

    vch =tuplaCVMCh[i][0]*tuplaCVMCh[i][1]
    vvh =tuplavh[i][0]*tuplavh[i][1]
    
    volumen_cvmch.append(vch)
    volumen_vh.append(vvh)
    
Tabla_Rt_CVMCh ['Volumen (km3)'] = volumen_cvmch  ### segun bibliografía el volumen en el presente es de 1.038 km3 (hay una diferencia de 0.286 km3)
Tabla_Rt_VH ['Volumen (km3)'] = volumen_vh
Tabla_Rt_CVMCh

print("*********************************************************************")
print("nombre variante = 'Tabla_Rt_CVMCh' ")
print("*********************************************************************")
print("RESULTADOS CVMCh")
print(Tabla_Rt_CVMCh)
print("*********************************************************************")
print("*********************************************************************")
print("nombre variante = 'Tabla_Rt_VH' ")
print("*********************************************************************")
print("RESULTADOS VH")
print(Tabla_Rt_VH)
print("*********************************************************************")


def save_excel_file(df, filename):
    if os.path.isfile(filename):
        print(f"El archivo '{filename}' ya existe aquí")
    else:
        df.to_excel(filename)
        print(f"El archivo '{filename}' se ha guardado")

# Guardar el archivo para CVMCh
filename_cvmch = "Output/Tabla_radiosfinal_CVMCh.xlsx"
save_excel_file(Tabla_Rt_CVMCh, filename_cvmch)

# Guardar el archivo para VH
filename_vh = "Output/Tabla_radiosfinal_VH.xlsx"
save_excel_file(Tabla_Rt_VH, filename_vh)

    
```
